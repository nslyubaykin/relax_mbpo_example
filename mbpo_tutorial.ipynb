# Defining Environment Dynamics Models

DeltaEnvModel works only with continuous environments and it uses an ensemble of neural networks to model environment dynamics in the form: 

$ \hat{\Delta}_{t+1}  = f_{\theta}(s_{t}, a_{t})$


$ \hat{s}_{t+1} = \hat{\Delta}_{t+1} + s_{t} $

Which means that neural net models the change in state given the current state and action. Model itself as an input takes a batch of observations with a shape (BATCH_SIZE, \*obs.shape) and a batch of corresponding actions with a shape (BATCH_SIZE, ACS_DIM) and outputting a batch of $ \hat{\Delta}_{t+1} $ with a shape (BATCH_SIZE, \*obs.shape) (same shape as observations batch).

In case of absence of the privileged information - explicit reward function, it also may be modeled by an ensemble of neural networks in the following form: 


$ \hat{r}_{t} = f_{\phi}(s_{t}, a_{t}) $

Reward model itself as an input takes a batch of observations with a shape (BATCH_SIZE, \*obs.shape) and a batch of corresponding actions with a shape (BATCH_SIZE, ACS_DIM) and outputting a batch of rewards with a shape (BATCH_SIZE, 1).

Therefore any custom user defined networks meeting these requirements should work well with DeltaEnvModel environment model.

So let's examine the default implementation of observations and rewards models offered by ReLAx:

# Define DeltaEnvModel

We will create a function for defining DeltaEnvModel to run it in a loop for multiple experiments. That is done in order to benchmark MBPO-SAC vs vanilla SAC

!Note: In practice, model-based acceleration still needs some privileged information, more precisely, termination function. For example, official implementation of MBPO [uses](https://github.com/jannerm/mbpo/blob/master/mbpo/static/hopper.py) pre-defined termination functions. That is done because without knowledge of it, when branching synthetic rollouts from real transitions, you may end up with some beyond-terminal transitions in training data. Obviously, as environment model has never seen any of such beyond-terminal transitions when training they tend to have very poor quality and usually cause training to diverge. In order to avoid it, termination function is used to cut off beyond-terminal transitions.

```python
def get_model(obs_dim, acs_dim, 
              terminal_function,
              ensemble_size=7):
    
    obs_models = []
    rews_models = []
    for i in range(ensemble_size): 
        obs_models.append(
            ContObsContAcsToObsMLP(input_obs_dim=obs_dim, output_obs_dim=obs_dim,
                                   acs_dim=acs_dim, nunits=250, nlayers=2) #350 for ant
        )
        rews_models.append(
            ContObsContAcsToRewsMLP(input_obs_dim=obs_dim, acs_dim=acs_dim, 
                                    nunits=250, nlayers=2)
        )
    
    # Creating schedules:
    # first 5000 iterations no learning - just data collection
    model_lr = PiecewiseSchedule({0: 5000}, 1e-3)

    model = DeltaEnvModel(
        obs_models=obs_models,
        device=torch.device('cuda'),
        learning_rate=model_lr,
        noise=0.01,
        batch_size=512,
        rews_models=rews_models,
        terminal_function=terminal_function,
        n_steps_per_update=1, 
        update_freq=10,
        stats_recalc_freq=500,
        random_agg=False,
    )
    
    return model
```

# Define SAC actor and CDQN critic

We will also wrap actor and critic definition in  a function to run it in a loop for multiple experiments. That is done in order to benchmark MBPO-SAC vs vanilla SAC

```python
def get_actor_and_critic(obs_dim, acs_dim,
                         n_steps_per_update=1):
    
    # Creating schedules:
    # first 10000 iterations no learning - just data collection
    policy_lr = PiecewiseSchedule({0: 10000}, 1e-3)
    critic_lr = PiecewiseSchedule({0: 10000}, 1e-3)
    
    actor = SAC(
        device=torch.device('cuda'),
        policy_net=TanhNormalMLP(obs_dim, acs_dim),
        auto_tune_alpha=True,
        target_entropy=-acs_dim,
        learning_rate=policy_lr,
        batch_size=100,
        n_random_steps=10000,
        n_steps_per_update=n_steps_per_update,
        min_acs=-1,
        max_acs=1
    )
    
    critic = CDQN(
        critic_net=ContQMLP(obs_dim, acs_dim),
        critic_net2=ContQMLP(obs_dim, acs_dim),
        device=torch.device('cuda'),
        learning_rate=critic_lr,
        weight_decay=0.0
    )

    return actor, critic
   
```

# Create training loop for comparison of MBPO-SAC vs vanilla SAC

Overall MBPO pipeline:

0) Initialize real-data replay buffer and synthetic-data replay buffer

1) Sample 1 transition from an environment with `SAC` actor

2) Add it to real-data replay buffer

3) Train obs models: $ \hat{\Delta}_{t+1}  = f_{\theta}(s_{t}, a_{t})$ and rewards models: $ \hat{r}_{t} = f_{\phi}(s_{t}, a_{t}) $ on real-data replay buffer

4) Sample a batch from real-data replay buffer

5) Simulate short N-step rollouts branched from a batch of real data with `DeltaEnvModel`. Use `SAC` actor for action generation. (On-practice mix it with the real data at the rate of ~5%)

6) Add simulated rollouts to synthetic-data replay buffer

7) Train `SAC` actor and `CDQN` critic using samples from synthetic-data replay buffer. (typically, number of training steps is significantly greater than 1)

## To accelerate replay buffer sample with ReLAx you should use `.accelerate()` method of `BufferSample`, which has the following args:

!Note: MBPO Model-Based acceleration also supported for all off-policy value-based algorithms.

1) `model`: an instance of environment model (`DeltaEnvModel` in our case)

2) `actor`: an actor to get actions for synthetic branched rollouts (`SAC` in our case)

3) `h`: the length of synthetic rollouts (0 for the first 20k steps (no acceleration) then gradually ascending to 5 (`PiecewiseSchedule`))

4) `tau`: the share of real transitions that will be used as a starting point for synthetic rollout (if > 1, then one transition will be used multiple times as a starting point, 1 in our case (also can be scheduled))

5) `real_ratio`: the share of real rollouts among generated ones. Mostly needed for MBPO rather than DYNA-Q (0.05 in our case)

6) `inplace`: if `True` - synthetic rollouts will be added to `BufferSample` itself and nothing will be returned. If `False` synthetic rollouts will be returned as a separate `PathList`. (Here: `False`)

7) `train_sampling`: sampling mode of actor (enables exploration and stochasticity, `True` in our case)

8) `cut_tails`: if starting real transition is N-th from the end of a sampled trajectory, then synthetic rollout generated from it won't be longer than N even if `h` is greater than N. Done to possibly avoid beyond-terminal transitions if termination function is unknown. (`False` in our case as we specify termination function)

```python
env_name = 'Hopper-v2'
terminal_function = hopper_v2_terminal_fn

acs_dim = 3
obs_dim = 11
n_experiments = 4
sac_n_steps = 1000000
mbpo_n_steps = 150000 # truncated to save time

logdir = '/home/nikita/Desktop/RelaxExamples/MBPO/tensorboard_logs'
sac_log_every = 10000
mbpo_log_every = 2000  # to have a better resolution
logs_by_run = {}

for exp_i in range(n_experiments):
    
    ######################
    # Running vanilla SAC#
    ######################
    print(f'Running vanilla SAC, experiment {exp_i}...')
    
    # Logging
    run_name = f'sac_{env_name}_experiment_{exp_i}'
    logger = SummaryWriter(log_dir=f'{logdir}/{run_name}')
    all_logs = []
    
    # Create training and eval environments
    env = gym.make(env_name)
    eval_env = gym.make(env_name)
    eval_env.seed(np.random.randint(100))
    
    # Actor & critic:
    actor, critic = get_actor_and_critic(obs_dim=obs_dim, acs_dim=acs_dim)
    
    # provide actor with critic
    actor.set_critic(critic)
    
    # Create replay buffer and wrap envs into Sampler
    sampler = Sampler(env=env)
    eval_sampler = Sampler(env=eval_env)
    replay_buffer = ReplayBuffer(int(1e+6))
    
    for step in tqdm(range(sac_n_steps)): 
    
        # sample training batch
        unit_pathlist = sampler.sample(actor=actor,
                                       n_transitions=1,
                                       max_path_length=None, 
                                       reset_when_not_done=False,
                                       train_sampling=True)

        # add transitions to replay buffer
        replay_buffer.add_paths(unit_pathlist)

        # update critic
        critic_logs = critic.update(replay_buffer)

        # update actor
        actor_logs = actor.update(replay_buffer)

        if step % sac_log_every == 0 and step > 0:

            # create logs of buffer
            buffer_logs = replay_buffer.create_logs(prefix='buffer')

            # evaluating the policy
            eval_batch = eval_sampler.sample_n_episodes(actor=actor,
                                                        n_episodes=5,
                                                        max_path_length=None, 
                                                        train_sampling=False)
            eval_logs = eval_batch.create_logs(prefix='eval')
            logs = {**critic_logs,
                    **actor_logs,
                    **buffer_logs,
                    **eval_logs}
            all_logs.append(logs)

            for key, value in logs.items():
                logger.add_scalar(key, value, step)

    logger.flush()
    
    # Saving run's logs
    logs_by_run[run_name] = all_logs
    
    # Saving models:
    actor.save_checkpoint('trained_models/sac', run_name+'_actor')
    critic.save_checkpoint('trained_models/sac', run_name+'_critic')
    
    ####################
    # Running MBPO SAC #
    ####################
    print(f'Running MBPO SAC, experiment {exp_i}...')
    
    # Logging
    run_name = f'mbpo_sac_{env_name}_experiment_{exp_i}'
    logger = SummaryWriter(log_dir=f'{logdir}/{run_name}')
    all_logs = []
    
    # Create training and eval environments
    env = gym.make(env_name)
    eval_env = gym.make(env_name)
    eval_env.seed(np.random.randint(100))
    
    # setting the schedules for acceleration
    # Synthetic rollouts horizon
    horizon_schedule1 = LinearSchedule(-1.3, 5, 100000, discrete=True)
    horizon_schedule2 = PiecewiseSchedule({0: 20001}, 1)
    horizon_schedule = CombinedSchedule(horizon_schedule1, horizon_schedule2, max)
    
    # Number of steps per MBPO update
    n_steps_per_update_schedule = PiecewiseSchedule({1: 20001,
                                                     5: 5000,
                                                     10: 5000}, 20)
    
    # Actor & critic:
    actor, critic = get_actor_and_critic(obs_dim=obs_dim, acs_dim=acs_dim,
                                         n_steps_per_update=n_steps_per_update_schedule)
    
    # Model
    model = get_model(obs_dim=obs_dim, acs_dim=acs_dim, 
                      terminal_function=terminal_function,
                      ensemble_size=7)
    
    # provide actor with critic
    actor.set_critic(critic)
    
    # Create replay buffers and wrap envs into Sampler
    sampler = Sampler(env=env)
    eval_sampler = Sampler(env=eval_env)
    replay_buffer = ReplayBuffer(int(1e+6))
    synthetic_buffer = ReplayBuffer(int(1e+6))
    
    for step in tqdm(mbpo_n_steps):

        # Collect training data
        unit_pathlist = sampler.sample(policy=policy,
                                       n_transitions=1,
                                       max_path_length=None, 
                                       reset_when_not_done=False,
                                       train_sampling=True)

        # add transitions to replay buffer
        replay_buffer.add_paths(unit_pathlist)

        # update_models on real data
        model_logs = model.update(replay_buffer)


        # add transitions to synthetic buffer
        horizon = horizon_schedule.value(step)
        horizon_logs = {}
        horizon_logs['mba_horizon'] = horizon
        
        # add transitions to synthetic buffer
        if step <= 20000:
            # just feed from the real data
            synthetic_buffer.add_paths(unit_pathlist.copy()) # creating an another copy of real data
            if step == 20000:
                synthetic_buffer.finish_paths() # mark all rollouts as finished
        else:
            # start MBA
            assert horizon > 0
            sample = replay_buffer.sample(400)
            synth_pathlist = sample.axcelerate(model=model,
                                               actor=actor,
                                               h=horizon,
                                               tau=1.,
                                               real_ratio=0.05,
                                               inplace=False,
                                               train_sampling=True,
                                               cut_tails=False)
            synthetic_buffer.add_paths(synth_pathlist)

        # update critic on synthetic data
        critic_logs = critic.update(synthetic_buffer)

        # update actor on synthetic data
        actor_logs = policy.update(synthetic_buffer)

        
        if step % mbpo_log_every == 0:

            # create logs of real buffer
            buffer_logs = replay_buffer.create_logs(
                prefix='buffer',
                extra_metrics={'training_time': (datetime.now() - train_start).seconds}
            )

            # create logs of synthetic buffer
            synth_buffer_logs = synthetic_buffer.create_logs(
                prefix='synth_buffer'
            )

            # evaluating the policy
            eval_batch = eval_sampler.sample_n_episodes(actor=actor,
                                                        n_episodes=5,
                                                        max_path_length=None, 
                                                        train_sampling=False)
            eval_logs = eval_batch.create_logs(prefix='eval')
            logs = {**model_logs,
                    **horizon_logs,
                    **critic_logs,
                    **actor_logs,
                    **buffer_logs,
                    **synth_buffer_logs,
                    **eval_logs}
            all_logs.append(logs)

            for key, value in logs.items():
                logger.add_scalar(key, value, step)

    logger.flush()
    
    # Saving run's logs
    logs_by_run[run_name] = all_logs
    
    # Saving models:
    actor.save_checkpoint('trained_models/mbpo_sac', run_name+'_actor')
    critic.save_checkpoint('trained_models/mbpo_sac', run_name+'_critic')
    model.save_checkpoint('trained_models/mbpo_sac', run_name+'_model')
    
```

# Visualize trained model's predictions:

Comparing actual obs & rews vs fitted obs & rews given actually happened actions input sequences:

```python
# formulate actual obs, starting obs, and actual action sequence
start_index = 150
end_index = 200

assert end_index < min([p.steps for p in batch.rollouts]) # ensure that rollouts are long enough

obs, rews, acs = [], [], []
for p in batch.rollouts:
    obs.append(p.data['obs'][start_index:end_index])
    acs.append(p.data['acs'][start_index:end_index])
    rews.append(p.data['rews'][start_index:end_index])
obs, rews, acs = np.array(obs), np.array(rews), np.array(acs)
obs, rews, acs = np.swapaxes(obs, 0, 1), np.swapaxes(rews, 0, 1), np.swapaxes(acs, 0, 1)

starting_obs = obs[0, :, :]
print(starting_obs.shape, obs.shape, rews.shape, acs.shape)

# Obtain predictions with a model
pred_obs, pred_rews, _ = model.predict_action_sequence(lag_obs=starting_obs,
                                                       action_sequence=acs)
print(pred_obs.shape, pred_rews.shape)
```

```python
# Calculate model's forecast MAE vs naive mean obs forecast MAE
mean_obs = batch.unpack(['obs']).mean(axis=0)[np.newaxis, np.newaxis, :]
mean_rews = batch.unpack(['rews']).mean(axis=0)

act_obs, act_rews = obs, rews

obs_mae = np.abs(act_obs - pred_obs)
mean_obs_mae = np.abs(act_obs - mean_obs)
rews_mae = np.abs(act_rews - pred_rews)
mean_rews_mae = np.abs(act_rews - mean_rews)
obs_mae = np.mean(obs_mae, axis=(1, 2))
mean_obs_mae = np.mean(mean_obs_mae, axis=(1, 2))
rews_mae = np.mean(rews_mae, axis=1)
mean_rews_mae = np.mean(mean_rews_mae, axis=1)
```

By looking at the graphs above we can conclude that learned dynamics model gives credible forecasts within the horizon of 10-20 steps ahead for this environment. After that, accumulated prediction error distorts forecasts too much. This is exactly the motivation to use only short (up to 5-10) branched rollouts for augmenting real data with imaginary rollouts. This logic is adopted both by MBPO and DYNA-Q.
